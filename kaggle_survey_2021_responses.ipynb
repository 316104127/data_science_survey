## Primeiros questionários

Time

```python
df.info()
```

```python
df.columns.values
```

```python
#Seleciono os primeiros questionários
col_select = ['Time from Start to Finish (seconds)', 'Q1', 'Q2', 'Q3', 'Q4', 'Q5', 'Q6']
df1 = df.filter(items=col_select)
df1.head()
```

```python
#Renomeio as colunas e tiro a primeira linha fora do padrão de dados.
df1 = df1.rename(columns={'Time from Start to Finish (seconds)':'Time_Seconds','Q1':'Age_Years','Q2':'Gender','Q3':'Country','Q4':'Formal_Education','Q5':'Title_Similar','Q6':'Writing_Code_Years'})
df1.drop([0], inplace=True)
df1.head()
```

```python
#Observo a tipagem das minhas colunas, vejo que a coluna Time_Seconds está como Objecto e mudo para Int, assim poderei fazer cálculos.
df1.info()
```

```python
#Faço a alteração, e divido por 60 para ter a coluna por minutos e verifico a tipagem dela. 
df1['Time_Seconds'] = df1['Time_Seconds'].astype(int)
df1['Time_Seconds'] = df1['Time_Seconds'] / 60
df1.rename(columns={'Time_Seconds':'Time_Min'}, inplace=True)
df1.info()
```

*Tempo **máximo** para realizar o questionário: 41477.55 minutos*

*Tempo **mínimo** para realizar o questionário: 2.00 minutos*

*Tempo **médio** geral para realizar o questionário: 184.24 minutos*


```python
df1.describe()
```

```python
#Filtro as colunas de Ano e tempo, para verificar a densidade do tempo para responder os questionários dividido pela idade
density_year_min = df1[['Age_Years','Time_Min']]
density_year_min.head()
```

Em uma média de 18 à 29 anos o tempo de demora é muito maior pois o número de registros é diferente dos demais, caso contrário a partir de 40 anos o tempo de demora seria maior.

```python
fig = px.density_heatmap(density_year_min,x='Age_Years', y='Time_Min', nbinsx=20, nbinsy=20, color_continuous_scale="Plasma", text_auto=True)
fig.show()
```

```python
df_local_time = df1[['Time_Min','Country']]
df_local_time.head()
```

```python
#Crio uma tabela com a quantidade de participantes dos países
df_local_country_qtd = df_local_time['Country'].value_counts().reset_index()
df_local_country_qtd.rename(columns={'index':'Country','Country':'QTD'}, inplace=True)

pd.set_option('display.max_rows', None)
df_local_country_qtd.style.set_table_styles(
    [{'selector':'tr:hover',
      'props':[('background-color','white'),('color','black')]}]
)
```
